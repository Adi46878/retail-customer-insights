```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
transactions = pd.read_csv(r"D:\AnalytixLabs\Case Studies\Python Case Study\Case Study 1 - Retail Case Study\Transactions.csv")
customer = pd.read_csv(r"D:\AnalytixLabs\Case Studies\Python Case Study\Case Study 1 - Retail Case Study\Customer.csv")
prod_cat_info = pd.read_csv(r"D:\AnalytixLabs\Case Studies\Python Case Study\Case Study 1 - Retail Case Study\prod_cat_info.csv")
```

```python
temp = pd.merge(left= transactions, right= customer, left_on= 'cust_id', right_on= 'customer_Id', how= 'inner')
```

```python
Customer_final = pd.merge(left= temp, right = prod_cat_info, left_on=['prod_cat_code', 'prod_subcat_code'], \
         right_on=['prod_cat_code', 'prod_sub_cat_code'], how= 'inner')
```
